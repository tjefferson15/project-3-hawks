```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
data_1 = pd.read_csv('/Users/tjeff/Desktop/CODING/project-3-hawks/Resources/StockX_Data_Contest.csv')
df = data_1.copy()
df.head()
```

```python
df.columns
```

```python
df['Sneaker Name'] = df ['Sneaker Name'].apply(lambda x: x.replace('-',' '))
df.head()
```

```python
df['Sale Price'] = df['Sale Price'].apply(lambda x: x.replace('$' , ''))
df ['Sale Price'] = df['Sale Price'].apply(lambda x: x.replace(',', ''))
df.head ()
```

```python
df['Retail Price'] = df['Retail Price'].apply(lambda x: x.replace ('$', ''))
df.head()
```

```python
df.to_csv ('Cleaned_Data.csv', index = False) 
```
